```python
import BioSimSpace as BSS
import MDAnalysis as MDA
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot graph for free ligands minimisation
Mini_free_pot = MDA.auxiliary.XVG.XVGReader("Minimisation/free_lig/potential.xvg")
x = []
y = []
for step in Mini_free_pot:
    y.append(step.data[1])
    x.append(step.time)
plt.title("Unbound State - Minimisation Energies")
plt.xlabel("Time [ps]")
plt.ylabel("Energy [KJ/mol]")
plt.plot(x, y, label="Potential")
plt.legend()
plt.savefig("Minimisation/free_lig/Free_Min_PotE")
```
